# add, sub, mul y div

```python
df2 = pd.DataFrame(np.random.randn(1000, 3), index=index,columns=['A', 'B', 'C'])

row = df.iloc[1]
column = df.A
print(df.head())
print('------------------')
print(row)
print('------------------')
print(column.head())
print('------------------')
print(df.add(column ,axis='index').head())
print('------------------')
print(df.add(column ,axis='rows').head())
print('------------------')
print((df + df2).head())
```

```python
print(df.head())
print('------------------')
print(row)
print('------------------')
print(column.head())
print('>------------------')
print(df[['B','C']].sub(column ,axis='index').head())
print('>>------------------')
print(df.sub(column ,axis='index').head())
print('------------------')
print((df - df2).head())
```

```python
print(df.head())
print('------------------')
print(row)
print('------------------')
print(column.head())
print('------------------')
print(df.div(column ,axis='index').head())
print('------------------')
print(df.div(row ,axis='columns').head())
print('------------------')
print((df / df2).head())
```
