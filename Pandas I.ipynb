# Regresion Lineal 



```python
#Regresion Lineal 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.cross_validation import train_test_split
```

```python
#Cargamos el conjunto de datos
dataset = pd.read_csv('heart.csv')
dataset.shape
```

```python
#dataset_rls = dataset[['age','trestbps']]
dataset_rls = dataset[['age','chol']]
```

```python
x = dataset_rls.iloc[:, :-1].values
y = dataset_rls.iloc[:, 1].values
```

```python
print(x.shape, y.shape)
```

```python
#dividimos los datos en el conjunto de entrenamiento y el conjunto de pruebas
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/2, random_state=0)
print(len(x_train), len(x_test), len(y_train), len(y_test))
```

```python
#Cargamos el conjunto de entrenamiento al modelo de Regresión Lineal
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
print(len(x_train), len(y_train))
```

```python
#Predicción de los resultados del conjunto de pruebas (x_test)
y_pred = regressor.predict(x_test)
len(y_pred)
#Ahora comparamos y_pred con los valores reales y_test
```

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_test, y_pred, color = 'blue')
plt.title('Edad vs Presion en Sangre (Conjunto de prueba)')
plt.xlabel('Edad')
plt.ylabel('Presion en sangre')
plt.show()
```

```python
rmse = (np.sqrt(mean_squared_error(y_test, y_pred)))
r2 = r2_score(y_test, y_pred)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
